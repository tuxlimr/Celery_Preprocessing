### Cumulative Returns, drawdowns etc.


```python
apple = yf.download('AAPL').Close.to_frame()
apple
```

```python
apple['d_returns'] = np.log(apple.div(apple.shift(1)))
```

```python
apple
```

```python
apple.dropna(inplace=True)

apple
```

```python
apple.d_returns.sum()
```

```python
np.exp(apple.d_returns.sum())
```

```python
apple['cumm_returns'] = apple.d_returns.cumsum().apply(np.exp)
apple
```

```python
apple.cumm_returns.plot(figsize=(12,8), title="Apple buy and Hold", fontsize=12)
plt.show()
```

```python
apple.d_returns.mean()*252
```

### So apple has given 17% average return on annual basis

```python
apple.d_returns.std()*np.sqrt(252)
```

### Calulate draw downs


```python
apple['cummmax']= apple.cumm_returns.cummax()
apple
```

```python
apple[['cumm_returns','cummmax']].plot(figsize=(12,8), title="Apple buy and Hold + cummmax", fontsize=12)
plt.show()
```

```python
apple['drawdown'] =apple['cummmax']- apple['cumm_returns']
```

```python
apple
```

```python
apple.drawdown.max()
```

### When did this drop down happened

#### SMA Strategy

```python
data = apple.Close.loc[(apple.index>='1991-01-01')]
data = data.to_frame()
data
```

```python
sma_s = 50
sma_l = 100
```

```python
data['sma_s'] = data.Close.rolling(sma_s).mean()
data['sma_l'] = data.Close.rolling(sma_l).mean()
data
```

```python
data.plot(figsize=(12,8), title= f"Apple{sma_s}|{sma_l} ",fontsize=12)
```

```python
data.loc['2016'].plot(figsize=(12,8), title= f"Apple{sma_s}|{sma_l} ",fontsize=12)
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data['position'] = np.where(data['sma_s']> data['sma_l'],1,-1)

data
```

```python
data.loc[:, ["sma_s", "sma_l","position"]].plot(figsize=(12,8), title= f"Apple{sma_s}|{sma_l} ",fontsize=12)
```

```python
data.loc[:, ["sma_s", "sma_l","position"]].plot(figsize=(12,8), title= f"Apple{sma_s}|{sma_l} ",fontsize=12, secondary_y="position")
```

```python
data.loc["2023", ["sma_s", "sma_l","position"]].plot(figsize=(12,8), title= f"Apple{sma_s}|{sma_l} ",fontsize=12, secondary_y="position")
```

```python
data["return_b&h"]= np.log(data.Close.div(data.Close.shift(1)))
data
```

```python
data["Strategy"]= data["return_b&h"]*data.position.shift(1)
data
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data[['return_b&h','Strategy']].sum()
```

#### This means buy and Hold have better returns than immidiate buy and short

```python
data[['return_b&h','Strategy']].sum().apply(np.exp)
```

```python
data[['return_b&h','Strategy']].std()*np.sqrt(252) ## Annual return
```
