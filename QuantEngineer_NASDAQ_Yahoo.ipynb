```python
aapl = close.AAPL.copy().to_frame()
aapl
```

```python
aapl.shift(periods=1)
```

```python
aapl["lag1"] = aapl.shift(periods=1)
```

```python
aapl
```

```python
aapl["Diff"]= aapl.AAPL.sub(aapl.lag1)
aapl["% change"] = aapl.AAPL.div(aapl.lag1)
```

```python
aapl
```

### You can use diff method instead of sub method used as earlier.

```python
aapl["Diff2"]= aapl.AAPL.diff(periods=1)
```
