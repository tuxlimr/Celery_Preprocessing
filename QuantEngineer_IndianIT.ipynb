```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ticker = ['TCS.NS','INFY.NS','LTIM.NS', 'WIPRO.NS','HCLTECH.NS']
```

#### Pull Data

```python
stocks = yf.download(ticker, start="2010-01-01", end="2024-01-01")
```

```python
stocks
```

```python
stocks.to_csv("Stocks.csv")
```

```python
stocks = pd.read_csv("Stocks.csv", header=[0,1], index_col=[0], parse_dates=[0])
stocks.head()
```

```python
stocks.columns
```

### Flat Index

```python
stocks.columns = stocks.columns.to_flat_index()
stocks.columns
```

```python
stocks
```

```python
stocks.columns = pd.MultiIndex.from_tuples(stocks.columns)
stocks.columns
```

```python
stocks
```

```python
stocks.loc[:, "Close"]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
close = stocks.loc[:, "Close"]
```

```python
close.plot(figsize=(15,8), fontsize=12)
```

```python
close.plot(figsize=(15,8), fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
normclose = close.div(close.iloc[0]).mul(100)
normclose
```

```python
normclose.plot(figsize=(15,8), fontsize=12)
plt.legend(fontsize=12)
plt.show()
```
