```python
import yfinance as yf
import pandas as pd

class StockAnalyzer:
    def __init__(self, ticker_symbol, lookback_days=90):
        self.ticker_symbol = ticker_symbol
        self.lookback_days = lookback_days

    def _fetch_stock_data(self, period):
        stock = yf.Ticker(self.ticker_symbol)
        data = stock.history(period=period)
        return data

    def _calculate_break(self, data, field):
        prev_value = data[field].iloc[0]
        today_value = data[field].iloc[1]
        return today_value > prev_value


    def analyze_stock(self):
        data_2d = self._fetch_stock_data("2d")
        data_5d = self._fetch_stock_data("5d")
        data_10d = self._fetch_stock_data("10d")

        support_level = data_2d['Low'].min()
        resistance_level = data_2d['High'].max()

        prev_close = data_2d['Close'].iloc[0]
        today_close = data_2d['Close'].iloc[1]

        breakout_detected = int(today_close > prev_close)
        breakdown_detected = int(today_close < prev_close)

        resistance_breakout_detected = int(data_5d['High'].iloc[1] > data_5d['High'].iloc[0])
        support_breakdown_detected = int(data_5d['Low'].iloc[1] < data_5d['Low'].iloc[0])

        difference = today_close - prev_close
        difference_percentage = (difference / prev_close) * 100 if prev_close != 0 else 0

        # Determine trend based on the difference
        if difference > 0:
            trend = 'Up'
        elif difference < 0:
            trend = 'Down'
        else:
            trend = 'No Change'

        # Calculate 10-day moving average
        ma_10d = data_10d['Close'].mean()

        # Check for 10-day moving average breakout
        ma_breakout_detected = int(today_close > ma_10d)

        # Track breakout and breakdown counts
        breakout_count = breakout_detected + resistance_breakout_detected + ma_breakout_detected
        breakdown_count = breakdown_detected + support_breakdown_detected + support_breakdown_detected

        # Determine buy signal based on breakout and breakdown counts
        if breakout_count == 3:
            buy_signal = 'Strong Buy'
        elif breakout_count == 2:
            buy_signal = 'Buy'
        elif breakdown_count == 3:
            buy_signal = 'Strong Sell'
        elif breakdown_count == 2:
            buy_signal = 'Sell'
        else:
            buy_signal = 'Neutral'

        return {
            'Stock': self.ticker_symbol,
            'Support Level': support_level,
            'Resistance Level': resistance_level,
            'Previous Day\'s Close': prev_close,
            'Today\'s Close': today_close,
            'Breakout Detected': breakout_detected,
            'Breakdown Detected': breakdown_detected,
            'Resistance Breakout Detected': resistance_breakout_detected,
            'Support Breakdown Detected': support_breakdown_detected,
            'Trend': trend,
            'Difference': difference,
            'Difference Percentage': difference_percentage,
            '10 DMA Breakout Detected': ma_breakout_detected,
            '10 DMA': ma_10d,
            'Buy Signal': buy_signal
        }


if __name__ == "__main__":
    ticker_lists = {
        'NiftyIT': ['INFY.NS', 'TCS.NS'],
        'NiftyMetal': ['HINDALCO.NS', 'TATASTEEL.NS', 'JINDALSTEL.NS', 'VEDL.NS', 'ICICIBANK.NS'],
        'NiftyTelecom': ['BHARTIARTL.NS', 'TATACOMM.NS', 'IDEA.NS'],
        'NityBank': ['HDFCBANK.NS', 'PNB.NS', 'SBIN.NS', 'ICICIBANK.NS', 'AXISBANK.NS', 'CANBK.NS', 'BANKBARODA.NS', 'INDUSINDBK.NS', 'RBLBANK.NS', 'MAHABANK.NS', 'J&KBANK.NS', 'IDBI.NS'],
        'NIftyFMCG': ['BAJAJCON.NS', 'ITC.NS', 'HINDUNILVR.NS', 'DABUR.NS', 'COLPAL.NS']
    }

    stock_data_list = []

    for category, stocks in ticker_lists.items():
        for stock_symbol in stocks:
            analyzer = StockAnalyzer(stock_symbol)
            stock_data = analyzer.analyze_stock()
            stock_data['Category'] = category
            stock_data_list.append(stock_data)

    df = pd.DataFrame(stock_data_list)
df
```
