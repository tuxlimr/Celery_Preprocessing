```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the stock symbol (TCS)
stock_symbol = "TCS"

# Fetch historical data for TCS stock
tcs_data = yf.download(stock_symbol, period="5y")  # You can adjust the period as needed

# Calculate the 200-day moving average
tcs_data['200_MA'] = tcs_data['Adj Close'].rolling(window=200).mean()

# Create a new column for buy and sell signals
tcs_data['Signal'] = 0  # Initialize the signal column with zeros

# Generate buy signals (1) when the closing price crosses above the 200-day MA
tcs_data.loc[tcs_data['Adj Close'] > tcs_data['200_MA'], 'Signal'] = 1

# Generate sell signals (-1) when the closing price crosses below the 200-day MA
tcs_data.loc[tcs_data['Adj Close'] < tcs_data['200_MA'], 'Signal'] = -1

# Plot the stock's adjusted close price, the 200-day moving average, and buy/sell signals
plt.figure(figsize=(12, 6))
plt.plot(tcs_data.index, tcs_data['Adj Close'], label='TCS Adj Close', color='blue')
plt.plot(tcs_data.index, tcs_data['200_MA'], label='200-day MA', color='red')

# Filter buy and sell signals
buy_signals = tcs_data[tcs_data['Signal'] == 1]
sell_signals = tcs_data[tcs_data['Signal'] == -1]

# Plot buy signals as green triangles (^)
plt.scatter(buy_signals.index, buy_signals['Adj Close'], marker='^', color='green', label='Buy Signal', alpha=1)

# Plot sell signals as red inverted triangles (v)
plt.scatter(sell_signals.index, sell_signals['Adj Close'], marker='v', color='red', label='Sell Signal', alpha=1)

plt.title(f'{stock_symbol} Stock Price with 200-day MA and Buy/Sell Signals')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

# Print the buy and sell signals
print("Buy Signals:")
print(buy_signals[['Adj Close', '200_MA']])
print("\nSell Signals:")
print(sell_signals[['Adj Close', '200_MA']])
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the stock symbol (TCS)
stock_symbol = "TCS"

# Fetch historical data for TCS stock
tcs_data = yf.download(stock_symbol, period="5y")  # You can adjust the period as needed

# Calculate the 90-day and 200-day moving averages
tcs_data['90_DMA'] = tcs_data['Adj Close'].rolling(window=90).mean()
tcs_data['200_MA'] = tcs_data['Adj Close'].rolling(window=200).mean()

# Create columns for buy and sell signals based on both moving averages
tcs_data['Signal_90_DMA'] = 0  # Initialize the 90 DMA signal column with zeros
tcs_data['Signal_200_MA'] = 0  # Initialize the 200 MA signal column with zeros

# Generate buy signals (1) when the 90 DMA crosses above the 200 MA
tcs_data.loc[tcs_data['90_DMA'] > tcs_data['200_MA'], 'Signal_90_DMA'] = 1

# Generate sell signals (-1) when the 90 DMA crosses below the 200 MA
tcs_data.loc[tcs_data['90_DMA'] < tcs_data['200_MA'], 'Signal_90_DMA'] = -1

# Plot the stock's adjusted close price, the 90 DMA, the 200 MA, and buy/sell signals for both
plt.figure(figsize=(12, 6))
plt.plot(tcs_data.index, tcs_data['Adj Close'], label='TCS Adj Close', color='blue')
plt.plot(tcs_data.index, tcs_data['90_DMA'], label='90-day DMA', color='orange')
plt.plot(tcs_data.index, tcs_data['200_MA'], label='200-day MA', color='red')

# Filter buy and sell signals for both 90 DMA and 200 MA
buy_signals_90_DMA = tcs_data[tcs_data['Signal_90_DMA'] == 1]
sell_signals_90_DMA = tcs_data[tcs_data['Signal_90_DMA'] == -1]

# Plot buy signals for 90 DMA as green triangles (^)
plt.scatter(buy_signals_90_DMA.index, buy_signals_90_DMA['Adj Close'], marker='^', color='green', label='Buy Signal (90 DMA)', alpha=1)

# Plot sell signals for 90 DMA as red inverted triangles (v)
plt.scatter(sell_signals_90_DMA.index, sell_signals_90_DMA['Adj Close'], marker='v', color='red', label='Sell Signal (90 DMA)', alpha=1)

plt.title(f'{stock_symbol} Stock Price with 90-day DMA and 200-day MA Crossovers')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

# Print the buy and sell signals for both 90 DMA and 200 MA
print("Buy Signals (90 DMA):")
print(buy_signals_90_DMA[['Adj Close', '90_DMA', '200_MA']])
print("\nSell Signals (90 DMA):")
print(sell_signals_90_DMA[['Adj Close', '90_DMA', '200_MA']])
```

### Create main function

```python
# @title
def main_function(stock="TCS"):
  # Fetch historical data for stock
  data = yf.download(stock, period="5y")  # You can adjust the period as needed

  # Calculate the 90-day and 200-day moving averages
  data['90_DMA'] = data['Close'].rolling(window=90).mean()
  data['200_MA'] = data['Close'].rolling(window=200).mean()

  # Create columns for buy and sell signals based on both moving averages
  data['Signal_90_DMA'] = 0  # Initialize the 90 DMA signal column with zeros
  data['Signal_200_MA'] = 0  # Initialize the 200 MA signal column with zeros

  # Generate buy signals (1) when the 90 DMA crosses above the 200 MA
  data.loc[data['90_DMA'] > data['200_MA'], 'Signal_90_DMA'] = 1

  # Generate sell signals (-1) when the 90 DMA crosses below the 200 MA
  data.loc[data['90_DMA'] < data['200_MA'], 'Signal_90_DMA'] = -1

  # Plot the stock's adjusted close price, the 90 DMA, the 200 MA, and buy/sell signals for both
  plt.figure(figsize=(12, 6))
  plt.plot(data.index, data['Adj Close'], label='TCS Adj Close', color='blue')
  plt.plot(data.index, data['90_DMA'], label='90-day DMA', color='orange')
  plt.plot(data.index, data['200_MA'], label='200-day MA', color='red')

  # Filter buy and sell signals for both 90 DMA and 200 MA
  buy_signals_90_DMA = data[data['Signal_90_DMA'] == 1]
  sell_signals_90_DMA = data[data['Signal_90_DMA'] == -1]

  # Plot buy signals for 90 DMA as green triangles (^)
  plt.scatter(buy_signals_90_DMA.index, buy_signals_90_DMA['Adj Close'], marker='^', color='green', label='Buy Signal (90 DMA)', alpha=1)

  # Plot sell signals for 90 DMA as red inverted triangles (v)
  plt.scatter(sell_signals_90_DMA.index, sell_signals_90_DMA['Adj Close'], marker='v', color='red', label='Sell Signal (90 DMA)', alpha=1)

  plt.title(f'{stock} Stock Price with 90-day DMA and 200-day MA Crossovers')
  plt.xlabel('Date')
  plt.ylabel('Price')
  plt.legend()
  plt.grid(True)

  # Show the plot
  plt.show()

  # Print the buy and sell signals for both 90 DMA and 200 MA
  print("Buy Signals (90 DMA):")
  print(buy_signals_90_DMA[['Adj Close', '90_DMA', '200_MA']])
  print("\nSell Signals (90 DMA):")
  print(sell_signals_90_DMA[['Adj Close', '90_DMA', '200_MA']])
```

#### Run for Multiple Stocks

```python
# @title
for i in NSE:
  main_function(i)
```
