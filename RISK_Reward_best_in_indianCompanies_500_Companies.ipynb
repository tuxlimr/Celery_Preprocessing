```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nifty_500 = ['360ONE.NS',
'ABB.NS',
'ACC.NS',
'AIAENG.NS',
'APLAPOLLO.NS',
'AUBANK.NS',
'AARTIDRUGS.NS',
]
```

```python
summary_list = []

for i in nifty_500:
    stocks = yf.download(i, start="2013-01-01", end="2024-01-01")
    close = stocks['Close']
    normclose = close.div(close.iloc[0]).mul(100)
    ret = close.pct_change().dropna()
    summary = ret.describe().T[['mean', 'std']]
    summary['Stock'] = i
    summary['mean'] = summary['mean'] * 252
    summary['std'] = summary['std'] * np.sqrt(252)
    summary_list.append(summary)

# Concatenate the summaries for all stocks into a single DataFrame
summary_df = pd.concat(summary_list)
df = pd.DataFrame(summary_df).reset_index()
# Extract the values for 'Stock', 'Mean', and 'Std'
stock_values = df[df['index'] == 'Stock']['Close'].values
mean_values = df[df['index'] == 'mean']['Close'].values
std_values = df[df['index'] == 'std']['Close'].values

# Create a new DataFrame with the desired structure
df_pivot = pd.DataFrame({
    'Stock': stock_values,
    'Mean': mean_values,
    'Std': std_values
})

df_pivot
```
